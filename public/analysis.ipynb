##For Interface

```python
filteredfactors = ['insured_sex_','auto_model_','auto_make_','insured_occupation_','insured_education_level_','policy_state_']
factorswithDummies = []
for factor in filteredfactors:
  factorswithDummies = factorswithDummies + list(coefficients.index[coefficients.index.str.startswith(factor)])
```

```python
factorswithDummies
```

```python
ourCoefficients = coefficients[coefficients.index.isin(factorswithDummies)]
```

```python
ourCoefficients
```

###Grouping Occupation

```python
from sklearn.cluster import KMeans

num_clusters = 8
occupationData = ourCoefficients[ourCoefficients.index.str.startswith('insured_occupation')]
df = pd.DataFrame(occupationData, columns=["Coefficients"])
kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init='auto')
df['Cluster'] = kmeans.fit_predict(df[['Coefficients']])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
for cluster_num in df['Cluster'].unique():
    cluster_data = df[df['Cluster'] == cluster_num]
    plt.scatter(cluster_data['Coefficients'], cluster_data.index, label=f'Cluster {cluster_num}', alpha=0.8)

plt.xlabel('Coefficients')
plt.title('K-Means Clustering')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.sort_values(by="Cluster")
```

```python
cluster_names = {
    0: "Machine Operations and Inspection",
    1: "Administrative and Clerical",
    2: "Handlers and Cleaners",
    3: "Sales or Other Services",
    4: "Armed Forces",
    5: "Protective and Technical Support",
    6: "Management Roles or Transportation",
    7: "Artisans and Professional Specialties"
}
df['Cluster'] = df['Cluster'].map(cluster_names)
```

```python
newOccupation = df.groupby('Cluster').mean()
newOccupation
```

###Replacing new with Old Occupation

```python
#get list of names of insured_occupation_
insured_occupation = ourCoefficients.index[ourCoefficients.index.str.startswith('insured_occupation')]
ourCoefficients.drop(insured_occupation, inplace=True)
newOccSeries = df.set_index('Cluster')['Coefficients']
```

```python
newOccSeries = newOccSeries.rename(lambda x: f"insured_occupation_{x}")
ourCoefficients = pd.concat([ourCoefficients, newOccSeries])

renameDict = {
    'policy_state_IL': 'policy_state_Illinois',
    'policy_state_IN':'policy_state_Indiana',
    'policy_state_OH':'policy_state_Ohio',
    'insured_sex_FEMALE': 'insured_sex_Female',
    'insured_sex_MALE': 'insured_sex_Male',
    'auto_make_Accura': 'auto_make_Acura',
    'auto_make_Suburu': 'auto_make_Subaru',
}

ourCoefficients.rename(index=renameDict , inplace=True)
```

```python
ourCoefficients
```

```python
ourWeights =  ourCoefficients / sum(abs(ourCoefficients))
```
